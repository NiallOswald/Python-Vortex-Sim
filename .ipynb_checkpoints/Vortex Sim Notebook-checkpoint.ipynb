## Vortex Street Simulation

```python
c = 4
f = lambda x, y: np.sqrt(((x**2 + y**2)**2 - x**2 + y**2)**2 + 4*(x**2)*(y**2))
dxdtf = lambda x, y: c*((x**2 + y**2)**2 - x**2 + y**2)/f(x, y) + 2
dydtf = lambda x, y: -c*(2*x*y)/f(x, y)

#dxdtf = lambda x, y: c
#dydtf = lambda x, y: 0

underlying_velocity = (dxdtf, dydtf)
```

```python
generation_points = ((0.1, 1), (0.1, -1))
```

```python
period = 40
```

```python
sim2 = VortexStreet(generation_points, period, underlying_velocity, dimensions = ((-2, -2), (10, 2)), circulation = 2*np.pi)
```

```python
sim2.save_sim(frames = 400, delta = 0.005, threshold = 10.0, gifdim = (12, 6))
```

```python
sim2.plot_pressure(((0, 1), (0, -1)), delta = 0.005, length = 200)
```
